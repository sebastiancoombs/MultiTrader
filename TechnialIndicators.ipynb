```python
import pandas as pd
from finta import TA
from IPython.display import display
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
source_data=pd.read_pickle('data/binanceus-BTCUSDT-1h.pkl')
data=source_data.copy()
```

```python
funcs=[c for c in dir(TA) if "__" not in c]
funcs
```

```python
def add_indicator(func_name,data):
    try:
        func=getattr(TA,func_name)
        indicator=func(data)
        # display(indicator.tail(5))

        try:
            col_name=f"feature_{indicator.name}".strip('.').replace(' ','_')
            data[col_name]=indicator.copy()
        except:
            cols=[f'feature_{func_name}_{n}'.strip('.').replace(' ','_') for n in indicator.columns.to_list()]
            data[cols]=indicator.copy()
    except Exception or FutureWarning as e:
        # print(f'{func_name} Didnt work')
        raise Exception
        
    return data.copy()
```

```python
working_funcs=[]
for func_name in funcs:
    try:
        data=add_indicator(func_name,data)
        working_funcs.append(func_name)
    except:
        print(f'{func_name} Didnt work')
```

```python

ta_data=source_data.copy()
```

```python
from tqdm.autonotebook import tqdm
```

```python
func_bar=tqdm(working_funcs)
for func_name in func_bar:
    try:
        func_bar.set_description(func_name)
        ta_data=add_indicator(func_name,ta_data)
        # working_funcs.append(func_name)
    except:
        print(f'{func_name} Didnt work')
```
